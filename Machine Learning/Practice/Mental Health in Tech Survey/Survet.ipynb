```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler, RobustScaler, Normalizer,OneHotEncoder, OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve, precision_recall_curve,f1_score
from sklearn.linear_model import LogisticRegression, LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor, GradientBoostingClassifier, GradientBoostingRegressor
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer,KNNImputer
warnings.filterwarnings("ignore")
df=pd.read_csv(r'G:\Python\Machine Learning\Practice\Mental Health in Tech Survey\survey.csv')
df
```

```python
#Chaning male to Male 
df['Gender'].replace('M', 'Male', inplace=True)
df['Gender'].replace('male', 'Male', inplace=True)
df['Gender'].replace('m', 'Male', inplace=True)
df['Gender'].replace('female', 'Female', inplace=True)
df['Gender'].replace('f', 'Female', inplace=True)
df['Gender'].replace('F', 'Female', inplace=True)

#drop other genders except male and female
df = df[df['Gender'].isin(['Male', 'Female'])]
df
```

```python
#predicitng missing values using simple imputer
cols=['comments','state','work_interfere','self_employed']
imputer = SimpleImputer(strategy='most_frequent')
df[cols] = imputer.fit_transform(df[cols])
df



    
```
