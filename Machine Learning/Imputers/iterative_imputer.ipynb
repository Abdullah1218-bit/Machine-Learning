# **Iterative Imputer**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plot
import numpy as np
import warnings
warnings.filterwarnings("ignore")
df=sns.load_dataset("titanic")

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer


# Select numeric columns for imputation (must not include non-numeric ones like 'sex')
data = df[['age', 'fare', 'pclass']]

# Create and apply the IterativeImputer
imputer = IterativeImputer(max_iter=10, random_state=0)
imputed_data = imputer.fit_transform(data)

# Replace the original columns with the imputed values
df[['age', 'fare', 'pclass']] = imputed_data
```
