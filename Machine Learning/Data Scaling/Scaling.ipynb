```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df=sns.load_dataset('tips')
df
```

```python
#Changing the categoric columns to numeric
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()
# Changing the categoric columns to numeric
oe.fit(df[['sex', 'smoker', 'day', 'time']])
df[['sex', 'smoker', 'day', 'time']] = oe.transform(df[['sex', 'smoker', 'day', 'time']])
```

# **Using Standard Scaler**

```python
#importing Standard Scalar
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler, Normalizer, MaxAbsScaler

scaler = StandardScaler()


# Fit the scaler to the data
scaler.fit(df[['total_bill', 'tip', 'sex', 'smoker', 'day', 'time', 'size']])

# Transform and replace the original columns
df[['total_bill', 'tip', 'sex', 'smoker', 'day', 'time', 'size']] = scaler.transform(
df[['total_bill', 'tip', 'sex', 'smoker', 'day', 'time', 'size']]
)

df

```

# *****MinMaxScaler*****

```python
df=sns.load_dataset('tips')
#Changing the categoric columns to numeric
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()
# Changing the categoric columns to numeric
oe.fit(df[['sex', 'smoker', 'day', 'time']])
df[['sex', 'smoker', 'day', 'time']] = oe.transform(df[['sex', 'smoker', 'day', 'time']])
df
```

```python
#Min Max Scalar
scaler = MinMaxScaler()

scaler.fit(df[['total_bill', 'tip', 'sex', 'smoker', 'day', 'time', 'size']])

df[['total_bill', 'tip', 'sex', 'smoker', 'day', 'time', 'size']] = scaler.transform(df[['total_bill', 'tip', 'sex', 'smoker', 'day', 'time', 'size']])

df
```

# ***Robust Scalar***

```python
df=sns.load_dataset('titanic')
df.dropna(inplace=True)
```

```python
# Using Ordinal Encoding changing all caterogic to numeric

from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder()

oe.fit(df[['sex', 'embarked', 'class','who','adult_male','embark_town','alone','deck']])

df[['sex', 'embarked', 'class','who','adult_male','embark_town','alone','deck']] = oe.transform(df[['sex', 'embarked', 'class','who','adult_male','embark_town','alone','deck']])

df

```

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()

scaler.fit(df[['sex', 'embarked', 'class','who','adult_male','embark_town','alone','deck']])

df[['sex', 'embarked', 'class','who','adult_male','embark_town','alone','deck']] = scaler.transform(df[['sex', 'embarked', 'class','who','adult_male','embark_town','alone','deck']])

df
```
