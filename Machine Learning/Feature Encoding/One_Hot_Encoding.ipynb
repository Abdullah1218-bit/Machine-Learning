# **One-Hot Encoding Using Pandas**

```python
# One Hot Encoding
df=pd.get_dummies(df,columns=["sex"])
df
```

```python
# change true and false to 0 and 1 
df["sex_female"]=df["sex_female"].map({True:1,False:0})
df["sex_male"]=df["sex_male"].map({True:1,False:0})
df

 
```

# **One-Hot Encoding Using Scikit-Learn**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
df=sns.load_dataset("titanic")
df
```

```python
from sklearn.preprocessing import OneHotEncoder


#One Hot Encoding
encoder=OneHotEncoder(sparse_output=False,handle_unknown='ignore')
cols_to_encode=["sex","embarked","class","who"]


# Fit and transform
encoded_array = encoder.fit_transform(df[cols_to_encode])


# Get new column names
encoded_columns = encoder.get_feature_names_out(cols_to_encode)


# Drop original columns
df.drop(columns=cols_to_encode, inplace=True)


# Add new columns
df[encoded_columns] = encoded_array  


# Show result
df
```
