# ***Label Encoding***

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=sns.load_dataset("titanic")
col_to_encode=["sex","embarked","class","who",'deck','embark_town',"alone",'adult_male']
#Importing Encoder from sk learn
from sklearn.preprocessing import LabelEncoder

#Creating a dictionary for storing the encoders of each column
Label_Encoder={}


#Looping through the columns to encode and encoding them one by one
for i in col_to_encode:
    le=LabelEncoder()
    df[i]=le.fit_transform(df[i])
    Label_Encoder[i]=le
df
```

```python
# Inverse transform the encoded columns
for i in Label_Encoder:
    df[i]=Label_Encoder[i].inverse_transform(df[i])
df
```
